```python
print("Signal length:",len(y),"samples")
print("Range of Magnitude:",np.min(y),"to", np.max(y))
print("Rate of Sampling:",sr, "Hz")
```

# A3. Take a small segment of the signal and play it.  

```python
# Dinitialising the start and end time
start_time = 2  
end_time = 3    

# Converting the time to indices samples
start_sample = int(start_time * sr)
end_sample = int(end_time * sr)

# Extracting the segment
segment = y[start_sample:end_sample]

# Visualization
plt.figure(figsize=(15, 5))
librosa.display.waveshow(segment, sr=sr)
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

ipd.Audio(segment, rate=sr)
```

# A4. Play around with your recorded speech signal for various segments. Understand the nature of the signal. Also observe with abruptly segmented speech, how the perception of that speech is affected. 

```python
# Set the number of segments to create
num_segments = 5

# Set the duration of each segment in seconds
segment_duration = 1

# list to store the segments
segments = []

for _ in range(num_segments):
    #intialising the max_start_time
    max_start_time = len(y) / sr - segment_duration
    
    # Generating the random start time
    start_time = np.random.uniform(0, max_start_time)
    
    # Converting the time to sample indices
    start_sample = int(start_time * sr)
    end_sample = int((start_time + segment_duration) * sr)
    
    # Extract and adding the segments in the list
    segments.append(y[start_sample:end_sample])

# Visualization
plt.figure(figsize=(15, 5))
for i, segment in enumerate(segments):
    plt.subplot(1, num_segments, i + 1)
    librosa.display.waveshow(segment, sr=sr)
    plt.title(f'Segment {i + 1}')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

for i, segment in enumerate(segments):
    print(f'Playing Segment {i + 1}')
    ipd.display(ipd.Audio(segment, rate=sr))
```
